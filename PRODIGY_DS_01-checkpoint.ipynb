```python
import pandas as pd
import matplotlib.pyplot as plt
import zipfile

zip_file_path = r'C:/Users/Anshuman/Downloads/API_SP.POP.TOTL_DS2_en_csv_v2_649149.zip'

with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    csv_file_name = 'API_SP.POP.TOTL_DS2_en_csv_v2_649149.csv'
    zip_ref.extract(csv_file_name, path='.')

csv_file_path = csv_file_name
population_data = pd.read_csv(csv_file_path, skiprows=4)

selected_countries = ['India', 'China', 'United Kingdom', 'United States', 'France', 'Sweden', 'Denmark', 'Netherlands', 'Norway']
filtered_data = population_data[population_data['Country Name'].isin(selected_countries)]

population_2023 = filtered_data[['Country Name', '2023']].dropna()

plt.figure(figsize=(12, 6))
plt.bar(population_2023['Country Name'], population_2023['2023'], color=['blue', 'red', 'green', 'orange', 'purple', 'cyan', 'magenta', 'yellow', 'black'])
plt.xlabel('Country')
plt.ylabel('Population in 2023')
plt.title('Population in 2023 for Selected Countries')
plt.xticks(rotation=45)
plt.show()
```
